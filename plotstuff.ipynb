```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # start, finish, n points
```

```python
d_r = 2.5*t
```

```python
d_s = 3*(t-5)
fix,ax = plt.subplots()
plt.title('A Bank Robber Caught in the Act')
plt.xlabel('Time (in minutes)')
plt.ylabel('Distance (in km)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='red')
plt.axvline(x=30,color='purple',linestyle='--')
_ = plt.axhline(y=75,color='blue',linestyle='--')
```
